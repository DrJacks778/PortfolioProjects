```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts configuration of plots

#read in data

df = pd.read_csv(r'C:\Users\Kickanannee Poonuggt\Downloads\movies.csv')
```

```python
df.head()
```

```python
#check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#column data types

df.dtypes
```

```python
df = df.dropna()
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# hunt duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Film Budget')

plt.show
```

```python
#correlation matrix

df.corr(numeric_only=True)
```

```python
#heatmap

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Figures')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#give company ident

df_numerized = df

for col_name in df_numerized.columns:
        if(df_numerized[col_name].dtype == 'object'):
            df_numerized[col_name] = df_numerized[col_name].astype('category')
            df_numerized[col_name] = df_numerized[col_name].cat.codes
            
df_numerized  
            
            
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Figures')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
